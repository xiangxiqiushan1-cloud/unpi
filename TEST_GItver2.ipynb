```python
import pandas as pd

# === CSVファイルを読み込み ===
df = pd.read_csv("finalkaiki.csv")

# === 数値データだけを対象に相関行列を計算（Pearson）===
corr_matrix = df.corr(method="pearson")

# === 結果をCSVに出力 ===
corr_matrix.to_csv("correlation_matrix.csv", encoding="utf-8-sig")

print("相関係数行列を correlation_matrix.csv に出力しました。")
```
